## Gráficos

```python
from math import ceil, log10
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from stemgraphic import stem_graphic

plt.style.use('bmh')
tab = pd.read_feather('./datasets/tab2_1.feather')

#Criando uma variável idade agregando anos e meses
tab['idade'] = tab['idade_anos'] + (tab['idade_meses']/12)

def create_freq_table(df, variable):
    values = df[variable].unique()
    
    abs_freq = [len(df[df[variable] == value]) for value in values]
    n = sum(abs_freq)
    
    rel_freq = [abs_freq_value / n for abs_freq_value in abs_freq]
    return pd.DataFrame({variable: values,
                         "Absolute Frequency": abs_freq,
                         "Relative Frequency": rel_freq}).sort_values(variable)

tab.head()
```

Além das tabelas de distribuição de frequência, uma maneira de entender como se distribuem os dados é por meio de gráficos. Existem diferentes formas de representar uma variável graficamente, dependendo do tipo de dado que estamos maipulando.

### Gráficos para variáveis qualitativas
1. Gráfico de barras

```python
freq_grau_instrucao = create_freq_table(tab, 'grau_instrucao')

plt.bar(['Fundamental', 'Médio', 'Superior'], freq_grau_instrucao['Absolute Frequency'])
plt.yticks([0, 5, 10, 15, 20])
plt.ylabel('Frequência')
plt.show()
```

2. Gráfico em setores (pizza)

```python
plt.pie(freq_grau_instrucao['Absolute Frequency'], labels=['Fundamental', 'Médio', 'Superior'], autopct='%1.1f%%')
plt.show()
```

### Gráficos para variáveis quantitativas
1. Gráfico de barras (variável discreta)

```python
n_filhos_min = tab['n_filhos'].min()
n_filhos_max = tab['n_filhos'].max()
labels = []
values = []
for n in np.arange(n_filhos_min, n_filhos_max+1):
    labels.append("{:1.0f}".format(n))
    values.append(tab[tab['n_filhos'] == n]['n_filhos'].count())

plt.bar(labels, values)
plt.show()
```

2. Gráfico de dispersão unidimensional (stripplot) (variável discreta)

```python
sns.stripplot(x=tab['n_filhos'], orient='h')
plt.show()
```

3. Histograma

```python
plt.hist(tab['salario'])
plt.show()
```

4. Ramo-e-folhas

```python
stem_graphic(tab['salario'])
plt.show()
```
